1. Write a program for Splay tree having functions for the following operations:
      
 Insert an element (no duplicates are allowed),
 
 Delete an element (existing as well as non-existing),

 Finding and deleting a minimum element,

 Finding and deleting a maximum element,

 Traverse the Splay tree (in-order, pre-order, and post-order),

 Splaying (Either bottom-up approach or top-down approach),

 Right rotation, and

 Left rotation.

```python
class Node:
    def __init__(self, key, left=None, right=None):
        self.key = key
        self.left = left
        self.right = right

class SplayTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root == None:
            self.root = Node(key)
            return

        self.root = self._splay(key, self.root)

        if self.root.key == key:
            return

        node = Node(key)

        if key < self.root.key:
            node.right = self.root
            node.left = self.root.left
            self.root.left = None
        else:
            node.left = self.root
            node.right = self.root.right
            self.root.right = None

        self.root = node

    def delete(self, key):
        if self.root == None:
            return

        self.root = self._splay(key, self.root)

        if self.root.key != key:
            return

        if self.root.left == None:
            self.root = self.root.right
        else:
            x = self.root.right
            self.root = self._splay(key, self.root.left)
            self.root.right = x

    def find_min(self):
        if self.root == None:
            return None

        node = self.root
        while node.left != None:
            node = node.left
        return self._splay(node.key, self.root)

    def find_max(self):
        if self.root == None:
            return None

        node = self.root
        while node.right != None:
            node = node.right
        return self._splay(node.key, self.root)
    
    def in_order(self, node=None):
        if node == None:
            node = self.root
        if node.left != None:
            self.in_order(node.left)
        print(node.key, end=" ")
        if node.right != None:
            self.in_order(node.right)

    def pre_order(self, node=None):
        if node == None:
            node = self.root
        print(node.key, end=" ")
        if node.left != None:
            self.pre_order(node.left)
        if node.right != None:
            self.pre_order(node.right)

    def post_order(self, node=None):
        if node == None:
            node = self.root
        if node.left != None:
            self.post_order(node.left)
        if node.right != None:
            self.post_order(node.right)
        print(node.key, end=" ")

    def right_rotate(self, node):
        left_child = node.left
        node.left = left_child.right
        left_child.right = node
        return left_child

    def left_rotate(self, node):
        right_child = node.right
        node.right = right_child.left
        right_child.left = node
        return right_child

    def _splay(self, key, node):
        if node == None:
            return None

        if key < node.key:
            if node.left == None:
                return node

            if key < node.left.key:
                node.left.left = self._splay(key, node.left.left)
                node = self.right_rotate(node)
            else:
                node.left.right = self._splay(key, node.left.right)
                if node.left.right != None:
                    node.left = self.left_rotate(node.left)

            if node.left == None:
                return node
            else:
                return self.right_rotate(node)
        else:
            if node.right == None:
                return node

            if key > node.right.key:
                node.right.right = self._splay(key, node.right.right)
                node = self.left_rotate(node)
            else:
                node.right.left = self._splay(key, node.right.left)
                if node.right.left != None:
                    node.right = self.right_rotate(node.right)

            if node.right == None:
                return node
            else:
                return self.left_rotate(node)

tree = SplayTree()
tree.insert(100)
tree.insert(50)
tree.insert(150)
tree.insert(25)
tree.insert(75)
tree.insert(125)
tree.insert(175)

tree.in_order()
```


2. Implement both the approaches of splaying, i.e top-down as well as bottom-up. For some sequence of inputs
analyze which approach is performing better.



Top-Down

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class SplayTree:
    def __init__(self):
        self.root = None
    
    def leftRotate(self, x):
        y = x.right
        x.right = y.left
        y.left = x
        return y
    
    def rightRotate(self, x):
        y = x.left
        x.left = y.right
        y.right = x
        return y
    
    def splay(self, key):
        if self.root is None:
            return
        
        dummy = Node(0)
        dummy.left = dummy.right = None
        l = r = dummy
        x = self.root
        
        while x is not None:
            if key < x.key:
                if x.left is None:
                    break
                if key < x.left.key:
                    x = self.rightRotate(x)
                    if x.left is None:
                        break
                r.left = x
                r = x
                x = x.left
            elif key > x.key:
                if x.right is None:
                    break
                if key > x.right.key:
                    x = self.leftRotate(x)
                    if x.right is None:
                        break
                l.right = x
                l = x
                x = x.right
            else:
                break
        
        l.right = x.left
        r.left = x.right
        x.left = dummy.right
        x.right = dummy.left
        self.root = x
    
    def insert(self, key):
        node = Node(key)
        if self.root is None:
            self.root = node
            return
        self.splay(key)
        if self.root.key == key:
            return
        if self.root.key < key:
            node.left = self.root
            node.right = self.root.right
            self.root.right = None
        else:
            node.right = self.root
            node.left = self.root.left
            self.root.left = None
        self.root = node
        
    def search(self, key):
        if self.root is None:
            return False
        self.splay(key)
        if self.root.key == key:
            return True
        return False

s = SplayTree()
s.insert(100)
s.insert(200)
s.insert(300)
print(s.search(200))
print(s.search(400)) 
```

Bottom-Up

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class SplayTree:
    def __init__(self):
        self.root = None
    
    def leftRotate(self, x):
        y = x.right
        x.right = y.left
        y.left = x
        return y
    
    def rightRotate(self, x):
        y = x.left
        x.left = y.right
        y.right = x
        return y
    
    def bottomUpSplay(self, key, root):
        if root is None:
            return None, None
        
        if key < root.key:
            root.left, node = self.bottomUpSplay(key, root.left)
            if root.left is None:
                return None, root
            root = self.rightRotate(root)
        elif key > root.key:
            root.right, node = self.bottomUpSplay(key, root.right)
            if root.right is None:
                return None, root
            root = self.leftRotate(root)
        else:
            node = root
            root = None
        return root, node
    
    def insert(self, key):
        node = Node(key)
        if self.root is None:
            self.root = node
            return
        self.root, n = self.bottomUpSplay(key, self.root)
        if n.key == key:
            return
        if n.key < key:
            node.left = n
            node.right = n.right
            n.right = None
        else:
            node.right = n
            node.left = n.left
            n.left = None
        self.root = node
        
    def search(self, key):
        if self.root is None:
            return False
        self.root, node = self.bottomUpSplay(key, self.root)
        if node is None or node.key != key:
            return False
        return True

s = SplayTree()
s.insert(100)
s.insert(200)
s.insert(300)
print(s.search(200))
print(s.search(400))
```

3. Extent the available function list for the following operations on a Splay tree:

 Split and

 Join

```python
class Node:
    def __init__(self, key, left=None, right=None):
        self.key = key
        self.left = left
        self.right = right

class SplayTree:
    def __init__(self):
        self.root = None
    
    def leftRotate(self, x):
        y = x.right
        x.right = y.left
        y.left = x
        return y
    
    def rightRotate(self, x):
        y = x.left
        x.left = y.right
        y.right = x
        return y
    
    def topDownSplay(self, key, root):
        if root is None:
            return None
        if key == root.key:
            return root
        if key < root.key:
            if root.left is None:
                return root
            if key < root.left.key:
                root.left.left = self.topDownSplay(key, root.left.left)
                root = self.rightRotate(root)
            elif key > root.left.key:
                root.left.right = self.topDownSplay(key, root.left.right)
                if root.left.right is not None:
                    root.left = self.leftRotate(root.left)
            if root.left is None:
                return root
            else:
                return self.rightRotate(root)
        else:
            if root.right is None:
                return root
            if key < root.right.key:
                root.right.left = self.topDownSplay(key, root.right.left)
                if root.right.left is not None:
                    root.right = self.rightRotate(root.right)
            elif key > root.right.key:
                root.right.right = self.topDownSplay(key, root.right.right)
                root = self.leftRotate(root)
            if root.right is None:
                return root
            else:
                return self.leftRotate(root)
    
    def insert(self, key):
        node = Node(key)
        if self.root is None:
            self.root = node
            return
        self.root = self.topDownSplay(key, self.root)
        if self.root.key == key:
            return
        if self.root.key < key:
            node.left = self.root
            node.right = self.root.right
            self.root.right = None
        else:
            node.right = self.root
            node.left = self.root.left
            self.root.left = None
        self.root = node
    
    def search(self, key):
        if self.root is None:
            return False
        self.root = self.topDownSplay(key, self.root)
        return self.root.key == key
    
    def split(self, key):
        if self.root is None:
            return None, None
        self.root = self.topDownSplay(key, self.root)
        if self.root.key <= key:
            leftTree = self.root
            rightTree = self.root.right
            leftTree.right = None
            return leftTree, rightTree
        else:
            rightTree = self.root
            leftTree = self.root.left
            rightTree.left = None
            return leftTree, rightTree

    def join(self, leftTree, rightTree):
        if leftTree is None:
            return rightTree
        if rightTree is None:
            return leftTree
        rightTree = self.topDownSplay(leftTree.key, rightTree)
        rightTree.left = leftTree
        return rightTree

    def inOrderTraversal(self, node):
        if node is None:
            return
        self.inOrderTraversal(node.left)
        print(node.key, end=" ")
        self.inOrderTraversal(node.right)

s = SplayTree()
s.insert(50)
s.insert(30)
s.insert(70)
s.insert(20)
s.insert(40)
s.insert(60)
s.insert(80)

leftTree = SplayTree()
rightTree = SplayTree()
leftTree.root, rightTree.root = s.split(60)

print("Left tree:", end=" ")
leftTree.inOrderTraversal(leftTree.root)
print("\nRight tree:", end=" ")
rightTree.inOrderTraversal(rightTree.root)

s = SplayTree()
s.root = s.join(leftTree.root, rightTree.root)

print("\nJoined tree:", end=" ")
s.inOrderTraversal(s.root)
```
